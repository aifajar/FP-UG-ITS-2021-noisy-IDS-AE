```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/aifajar/FP-UG-ITS-2021-noisy-IDS-AE/main/NUSW-NB15_GT.csv'

df = pd.read_csv(url)

df.head()
```

```python
df_tcp = df[df['Protocol']=='tcp']

df_tcp.head()
```

```python
df_attack_tuple = df_tcp[['Attack category', 'Attack subcategory', 'Source IP', 'Source Port', 'Destination IP', 'Destination Port']]

df_attack_tuple.head()
```

```python
df_attack_tuple.info()
```

```python
df_attack_tuple = df_attack_tuple.rename({'Attack category': 'atk_category', 'Attack subcategory': 'atk_subcategory', 'Source IP': 'src_ip', 
                                          'Source Port': 'src_port', 'Destination IP': 'dst_ip', 'Destination Port': 'dst_port'}, axis=1)

df_attack_tuple.head()
```

```python
df_attack_tuple.to_csv('clean_testing_dataset.csv', index=False)
```
